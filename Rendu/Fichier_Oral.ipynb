## 1. Nettoyage du jeu de données

Etape préliminaire mais qui a pris du temps : jeu de données de 14 252 établissements proposés sur Parcoursup. 118 variables, dont une grande majorité peu utiles en pratique pour l'analyse. Beaucoup de NaN, on a donc supprimé les variables qui en contenaient trop et mis les autres à la médiane. On a enlevé pas mal de variables jusqu'à obtenir un dataset de 31 variables, dont 7 qualitatives et 24 quantitatives.

## 3. ACP, MCA, FAMD, LDA

### 3.1. Analyse en composantes principales

- réduction de variance : la première dimension explique plus d'un quart de la variance du jeu de données. En représentant les variables dans les deux premiers plans factoriels, on voit que 3 variables sont logiquement très corrélées : part_mention_TB, part_mention_haute et score_mention. Les deux dernières sont des features qui ont été créées à partir de la première, donc l'ACP fait apparaître la redondance de ces variables. Sinon, l'ACP ne permet pas vraiment de faire apparaître une structure particulière des données et ne réduit pas beaucoup la dimension du dataset. 

Néanmoins, on peut utiliser différents coloriages en fonction notamment de svariables qualitatives pour tenter d'interpréter à quoi correspond chaque axe.
- L'axe 1 semble séparer les formations sélectives des non-sélectives.
- L'axe 2 sépare les petits établissements à gauche (peu de candidats) des plus grosses structures. Sépare aussi bien bacs généraux (à droite) de fillières technologiques ou pros.
- Beaucoup d'outliers en haut à droite des deux premiers plans : établissements publics, majoritairement non-sélectifs, avec beaucoup de candidats, bac généraux et dossiers "moyens" (peu de mentions TB). Comme on pouvait s'y attendre, cela correspond principalement à des licences.
- Cluster de CPGE très identifiable: en bas à droite. Pareil pour BTS. PASS et écoles d'ingé plutôt au milieu à droite
- Certaines formations très éparses, pas bien discriminées par les premiers plans de l'ACP. Ex: écoles de commerce, licences...

## 4. Méthodes de clustering

### 4.1. Clustering sur l'ACM

Synthèse des trois approches de clustering sur l'ACM (K-means, CAH, GMM)

Au total, les trois méthodes convergent vers une lecture très cohérente de l'espace des formations Parcoursup :

1. un pôle **universitaire** centré sur les **Licences**, **PASS** et **Licence_LAS**, plutôt **non sélectif**, public et de **grande capacité** ;
2. un pôle **technico-professionnel sélectif** dominé par les **BTS** et, dans une moindre mesure, les **BUT** ;
3. un pôle **très sélectif et très tendu**, porté par les **CPGE**, les **écoles d'ingénieur** et les **écoles de commerce**.

Ce résultat est important car il relie directement le clustering à l'interprétation de l'ACM :
- l'**axe 1** séparait déjà le monde des **Licences** de celui des formations **sélectives** ;
- l'**axe 2** mettait en avant un gradient de **tension académique** ;
- les clusters ne font donc que **stabiliser** et **rendre plus lisibles** ces oppositions.
